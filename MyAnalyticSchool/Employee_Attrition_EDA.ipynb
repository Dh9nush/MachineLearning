# MODEL Building

# importing necssary libraries

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
df= pd.read_csv('Employee-turnover.csv')
df.head()
```

# outlier Treatment

```python
df['Monthly Income (USD)']= np.log(df['Monthly Income (USD)'])
```

# Dropping Unnecssary columns

```python
df.drop(['Employee Count','Employee ID'],inplace=True,axis=1)
```

# Encoding Data

```python
colm=['Education','Job Involvement','Work Life Balance','Relationship Satisfaction','Environment Satisfaction','Job Satisfaction']
LE=LabelEncoder()
df_tran=df[colm].apply(LE.fit_transform)
df=pd.concat((df.drop(colm,axis=1),df_tran),axis=1)
```

```python
colm2=['Department', 'Job Role', 'Gender','Marital Status',
       'Education Field', 'Business Travel','Over Time','Attrition (Yes/No)']
df_dummy=pd.get_dummies(df[colm2])
df=pd.concat((df_dummy,df.drop(colm2,axis=1)),axis=1)
```

```python
df.drop(['Attrition (Yes/No)_No','Over Time_No'],inplace=True,axis=1)
```

```python
color=sns.color_palette("Greens",10)
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(),cmap=color)
```

- Attrition has heavy corelation with Age,OverTime and Work Life Balance

```python
X=df.drop('Attrition (Yes/No)_Yes',axis=1)
Y=df['Attrition (Yes/No)_Yes']
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=42)
```

```python
ss=StandardScaler()
xtrain=ss.fit_transform(xtrain)
xtest=ss.transform(xtest)
```

```python
xtrain=pd.DataFrame(xtrain,columns=X.columns)
xtest=pd.DataFrame(xtest,columns=X.columns)
```

```python
LR=LogisticRegression()
DC=DecisionTreeClassifier()
KN=KNeighborsClassifier()
SV=SVC()
RF=RandomForestClassifier()
```

```python
model=LR.fit(xtrain,ytrain)
LR_score=model.score(xtest,ytest)
model2=DC.fit(xtrain,ytrain)
DC_score=model2.score(xtest,ytest)
model3=KN.fit(xtrain,ytrain)
KN_score=model3.score(xtest,ytest)
model4=SV.fit(xtrain,ytrain)
sv_score=model4.score(xtest,ytest)
Model5=RF.fit(xtrain,ytrain)
RF_score=Model5.score(xtest,ytest)
```

```python
plt.plot(['LR_score','DC_score','KN_score','sv_score','RF_SCORE'],[LR_score,DC_score,KN_score,sv_score,RF_score],)
```

- By default Parameter we can see SVM and RandomForest  are performing best

# Changing max depth for tree


```python
DC=DecisionTreeClassifier(max_depth=2)
model2=DC.fit(xtrain,ytrain)
DC_score=model2.score(xtest,ytest)
```

```python
plt.plot(['LR_score','DC_score','KN_score','sv_score','RF_SCORE'],[LR_score,DC_score,KN_score,sv_score,RF_score],)
```

- As Expected Tree models Are performing Well With Data has there are factors with non linearity
